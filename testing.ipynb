```python
import pandas as pd
```

```python
df = pd.read_csv("registers\-- 2023-09-12 20-26-45-418575.csv")
```

```python
def get_next_move(row):

    moves_values = {
        "UP": {"LEFT": "LEFT", "UP": "FRONT", "RIGHT": "RIGHT"},
        "RIGHT": {"UP": "LEFT", "RIGHT": "FRONT", "DOWN": "RIGHT"},
        "DOWN": {"RIGHT": "LEFT", "DOWN": "FRONT", "LEFT": "RIGHT"},
        "LEFT": {"DOWN": "LEFT", "LEFT": "FRONT", "UP": "RIGHT"}
    }

    return moves_values[row["snake_dir"]][row["snake_next_dir"]]

    

x = df[["rel_hor_dist_obj", "rel_vert_dist_obj"]]
y = df.apply(get_next_move, axis = 1)

```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x, y)
```

```python
tree.plot_tree(clf)
```
